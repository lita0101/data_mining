```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('Salary_Data.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.plot(kind='scatter',x='YearsExperience',y='Salary')
plt.show()
```

```python
#membuat train dan test
x_train,x_test,y_train,y_test = train_test_split(dataset['YearsExperience'], dataset['Salary'])
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
plt.scatter(x_train, y_train, label='sample training', color='#5a7bb5')
plt.scatter(x_test, y_test, label='sample testing', color='#70def1', alpha=0.7)
plt.legend()
plt.show()
```

```python
#membuat models
LR = LinearRegression()
LR.fit(x_train.values.reshape(-1,1),y_train.values)
```

```python
#membuat prediksi
prediction = LR.predict(x_test.values.reshape(-1,1))
prediction
```

```python
plt.plot(x_test, prediction, label='Garis Linear Regression', color='r')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, label ='Hasil Data Test', color = '#70def1')
plt.legend()
plt.show()
```

```python
plt.plot(x_test, prediction, label='Garis Linear Regression', color='r')
plt.scatter(x_test, y_test, label ='Hasil Data Test', color = '#70def1')
plt.xlabel("Pengalaman Bekerja (Tahun)")
plt.ylabel('Jumlah Penjualan (Dollar)')
plt.legend()
plt.show()
```
